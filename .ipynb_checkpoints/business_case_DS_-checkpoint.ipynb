### Q2: Voer een initiële analyse uit op de dataset. Welke eerste algemene informatie haal je uit de dataset? Denk hierbij ook aan missende waarden, duplicaten en data types. Vervang alle missende waarden door lege strings.

```python
## Plaats jouw code tussen de hekjes [initiële analyse]
rows, cols = netflix_df.shape

# checking the datatypes
print("Dataframe Datatypes per Column")
print(netflix_df.dtypes)
print()

# checking for unique show_id
show_id_len = len(netflix_df["show_id"])
show_id_len_uni = len(netflix_df["show_id"].unique())
print(f"All show_id's are unique: {show_id_len == show_id_len_uni}\n")

# checking for unique title
show_title_len = len(netflix_df["title"])
show_title_len_uni = len(netflix_df["title"].unique())
print(f"All show titles are unique: {show_title_len == show_title_len_uni}\n")

# check for each column the missing values, based on NAN values
df_cols = netflix_df.columns
miss_lst = []
for i in df_cols:
    rows_count = netflix_df[f"{i}"].count()
    # print(len(netflix_df[f"{i}"].unique()))
    miss_val = rows - rows_count
    miss_lst.append(miss_val)
    if miss_val > 0:
        print(f"{i} has {rows_count} values, missing {miss_val} values")

fig, ax = plt.subplots(figsize=(10, 3))
ax.barh(df_cols, miss_lst, label="missing values")
for bar in ax.containers:
    if bar != 0:
        ax.bar_label(bar, padding=10, color="r")
    ax.bar_label(bar, padding=10)

plt.title("Missing Values per Column")
plt.xlabel("Missing Value Count")
ax.set_xlim(right=3000)
ax.legend()
ax.grid(alpha=0.2)
##
```

```python
## Plaats jouw code tussen de hekjes [missende waarden vervangen]
# check for already empty strings, if no empty strings, nothing will print
empty_check = netflix_df.replace('', np.nan).count()
for i in range(len(empty_check)):
    if empty_check[i] + miss_lst[i] == rows:
        continue
    empty_strings = rows - empty_check[i] - miss_lst[i]
    print(f"{empty_strings} empty strings found for {df_cols[i]}")

netflix_df_emp = netflix_df.replace(np.nan, "")
display(netflix_df_emp.head(10))
##
```

### Q3: In de 'rating' kolom staan 3 afwijkende waarden. Lokaliseer deze waarden en verwijder vervolgens de rijen die deze waarden bevatten. Laat de lege strings staan.

```python
## Plaats jouw code tussen de hekjes [lokaliseren en verwijderen afwijkende waarden]
rating_lst = netflix_df_emp["rating"].unique()
print(f"Unique values: {rating_lst}\n")

print("Unique value count")
print(netflix_df_emp["rating"].value_counts())
print()

# removing the rows with wrong values
for rating in rating_lst:
    if "min" in rating:
        netflix_df_emp.drop(netflix_df_emp[netflix_df_emp["rating"] == rating].index, inplace = True)
        
rating_filtered_lst = netflix_df_emp["rating"].unique()
# print(rating_filtered_lst)

print(f"removed rows, where rating is: {[x for x in rating_lst if x not in rating_filtered_lst]}")
##
```

### Q4: De waarden van een aantal kolommen staan nog in onhandige formats. Schrijf voor de volgende vragen code zodat de kolommen zich beter lenen voor visualisatie.

```python
# Q4.1: De kolommen 'title', 'cast', 'country', 'director' en 'description' worden niet gebruikt in deze opdracht.
#       Verwijder deze.

## Plaats jouw code tussen de hekjes
col_del = ["title", "cast", "country", "director", "description"]
netflix_df_drop = netflix_df_emp.drop(col_del, axis=1)
netflix_df_drop
##
```

```python
# Q4.2: De kolom 'date_added' staat in een onhandig format voor analyse (Month DD, YYYY). Maak twee nieuwe kolommen 
#       (year_added en month_added) die respectievelijk enkel het jaar en de maand bevatten van de kolom 'date_added'.

## Plaats jouw code tussen de hekjes
split_date_year = netflix_df_drop["date_added"].str.split(pat=",", expand=True)
split_month_day = split_date_year[0].str.split(pat=" ", expand=True)

netflix_df_drop["month_added"] = split_month_day[0]
netflix_df_drop["date_added"] = pd.to_datetime(netflix_df_drop["date_added"])
netflix_df_drop["year_added"] = pd.DatetimeIndex(netflix_df_drop["date_added"]).year.astype("Int64")

netflix_df_drop = netflix_df_drop.drop("date_added", axis=1)
netflix_df_drop
##
```

```python
# Q4.3: De kolom 'listed_in' bevat in veel gevallen meer dan 1 waarde. Dit is onhandig voor de visualisatie.
#       Maak daarom een nieuwe kolom (first_genre) aan in de netflix_df die voor de originele kolom alleen 
#       de eerste waarde bevat. Doe dit door de strings te splitsen (.split methode).

## Plaats jouw code tussen de hekjes
listed_lst = netflix_df_drop["listed_in"].str.split(pat=",", expand=True)
netflix_df_drop["first_genre"] = listed_lst[0]
netflix_df_drop = netflix_df_drop.drop("listed_in", axis=1)
netflix_df_drop
##
```

```python
# Q4.4: De kolom 'rating' bevat informatie over leeftijdsgebonden content van Netflix. Deze kijkcijfer codes zijn echter 
#       niet heel betekenisvol. Maak een dictionary (rating_audience) aan met keys (ratings) en values (labels) o.b.v. 
#       van de onderstaande codering. Zorg er tot slot voor dat 'rating_audience' een nieuwe kolom wordt in de netflix_df.


#1.Everyone (‘TV-G’, ‘G’)
#2.Younger kids (‘TV-Y’)
#3.Older kids (‘TV-Y7’, ‘TV-Y7-FV’, ‘TV-PG’, ‘PG’)
#4.Teens (‘TV-14’, ‘PG-13’)
#5.Adults (‘TV-MA’, ‘R’, ‘NC-17’)
#6.Unknown (‘UR’, ‘NR’)

## Plaats jouw code tussen de hekjes
rating_dict = {
    "TV-G": "Everyone",
    "G": "Everyone",
    "TV-Y": "Younger kids",
    "TV-Y7": "Older kids",
    "TV-Y7-FV": "Older kids",
    "TV-PG": "Older kids",
    "PG": "Older kids",
    "TV-14": "Teens",
    "PG-13": "Teens",
    "TV-MA": "Adults",
    "R": "Adults",
    "NC-17": "Adults",
    "UR": "Unknown",
    "NR": "Unknown"
}

print(rating_dict)

netflix_df_drop["rating_audience"] = netflix_df_drop["rating"].map(rating_dict)
netflix_df_drop = netflix_df_drop.drop("rating", axis=1)
netflix_df_drop
##
```

```python
# Q5.2: Gebruik matplotlib om een piechart te maken die het percentage films en tv series laat zien.
#       Geef de plot de volgende titel:'Content type distribution' en een fontsize van 15.

## Plaats jouw code tussen de hekjes
labels = "Movies", "Shows"
sizes = [n_movies, n_shows]

def func(pct, allvals):
    absolute = round(pct/100.*sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# Plotting
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(sizes, labels=labels, autopct=lambda pct: func(pct, sizes))

plt.title("Content type distribution", fontsize=15)
plt.tight_layout()

plt.savefig(curr_dir+"/figures/"+"content_distribution.png");
##
```

# Title Recommendations Based on Clusters

Titles that can be recommended to the customer based on the cluster and similarity.

Clusters are created based on the following features:
- Type: the type of content, movie or TV show
- Genre: the main genre of the content
- Audience: the target group the content is aimed at

## Cluster model type
KModes is used to classify the content as the data is non-numercial.

```python
# installation or update of needed library KModes
# !pip install kmodes

# features needed for cluster classification
ml_df = netflix_df_drop[["type", "first_genre", "rating_audience"]]

# removal of all nan-values
ml_df = ml_df.dropna()

display(ml_df.head(5))
print(f"Dataframe of the shape: {ml_df.shape}")
```

## Data Clustering and Hyperparameter tuning
Hyperparameter n_clusters, amout of clusters is based on the from kmodes.kmodes import KModes

#### n_clusters
What it is: The number of clusters to form as well as the number of centroids to generate.

What it does: Defines how many different clusters the algorithm will try to partition your data into. This is equivalent to the "k" in k-means or k-modes.

Common Values: Integer values greater than 1. The optimal number usually depends on the specific use case and can be determined using methods like the Elbow Method, Silhouette Analysis, etc.

#### init
What it is: Method for initialization.

What it does: Defines the way initial centroids are chosen for the clustering. Initialization methods can impact the convergence and quality of the clustering.

Common Values:

'Huang': A method proposed by Huang in his original paper on K-Modes.
'Cao': Another initialization method designed to produce good-quality initial modes, proposed by Cao et al.
'random': Choose 'k' observations (rows) at random from data for the initial centroids.

#### n_init
What it is: Number of time the k-modes algorithm will be run with different centroid seeds.

What it does: Determines how many times the clustering will be performed with different initial centroids. The final results will be the best output (in terms of cost) of the n_init consecutive runs.

Common Values: Integer values greater than or equal to 1. Typically set to 10 or more for a more robust result at the expense of computational time.

#### verbose
What it is: Verbosity mode.

What it does: Defines the level of detail of logs that will be printed to the console during the algorithm's execution. Useful for debugging or understanding the algorithm's steps.

Common Values:

0: No output
1: More detailed output, including progress and cost at each iteration.

```python
from kmodes.kmodes import KModes

costs = []
for num_clusters in range(1, 25+1): # Range of cluster numbers to test
    km = KModes(n_clusters=num_clusters, init='Huang', n_init=10, verbose=0)
    km.fit_predict(ml_df)
    costs.append(km.cost_)
```

```python
plt.plot(range(1, 25+1), costs, marker='o')
plt.title('Elbow Method For Optimal k')
plt.axvline(x=6, linestyle="--", color='r', label="6 clusters")
plt.axvline(x=10, linestyle="--", color='k', label="10 clusters")
plt.xticks(range(1, 25+1))
plt.xlabel('Number of clusters')
plt.ylabel('Cost')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```

```python
# Initialize K-Modes
km = KModes(n_clusters=10, init='Huang', n_init=5, verbose=0)

# Fit the clusters
clusters = km.fit_predict(ml_df)

# Add the cluster labels back to your DataFrame
ml_df['Cluster'] = clusters
```

```python
ml_df["Cluster"].value_counts()
```

```python
netflix_df_classified = netflix_df_drop.join(ml_df["Cluster"]).dropna()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x_categories = list(netflix_df_classified["type"].unique())
y_categories = list(netflix_df_classified["first_genre"].unique())
z_categories = list(netflix_df_classified["rating_audience"].unique())
x_values = netflix_df_classified["type"].to_list()
y_values = netflix_df_classified["first_genre"].to_list()
z_values = netflix_df_classified["rating_audience"].to_list()

# Map string categories to integers
x_int_values = [x_categories.index(val) for val in x_values]
y_int_values = [y_categories.index(val) for val in y_values]
z_int_values = [z_categories.index(val) for val in z_values]

color_values = netflix_df_classified["Cluster"].to_list()

# Create 3D scatter plot
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_int_values, y_int_values, z_int_values, c=color_values, cmap='jet')

# Set ticks and labels
ax.set_xticks(range(len(x_categories)))
ax.set_xticklabels(x_categories, ha='right')
ax.set_yticks(range(len(y_categories)))
ax.set_yticklabels(y_categories, ha='left', size=7)
ax.set_zticks(range(len(z_categories)))
ax.set_zticklabels(z_categories, ha='left')

# ax.set_xlabel('Type', labelpad=20, loc="right")
# ax.set_ylabel('Genre', labelpad=70)
# ax.set_zlabel('Target Group', labelpad=50)

plt.title("3D Visualisation")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

netflix_df_classified_movie = netflix_df_classified#[netflix_df_classified["type"] == "TV Show"]

# Sample data
x_categories = list(netflix_df_classified_movie["rating_audience"].unique())
y_categories = list(netflix_df_classified_movie["first_genre"].unique())
x_values = netflix_df_classified_movie["rating_audience"].to_list()
y_values = netflix_df_classified_movie["first_genre"].to_list()
color_values = netflix_df_classified_movie["Cluster"].to_list()

# Map string categories to integers
x_int_values = [x_categories.index(val) for val in x_values]
y_int_values = [y_categories.index(val) for val in y_values]

# Define a list of distinct colors. You can extend this list as needed.
color_list = ['r', 'g', 'b', 'c', 'm', 'y', 'k', '#FF5733', '#33FF57', '#3357FF']

# Create an empty scatter plot
fig, ax = plt.subplots(figsize=(8, 8))

# Get unique cluster numbers
unique_clusters = np.unique(color_values)

# Loop through unique cluster numbers and plot each cluster separately
for i, cluster in enumerate(unique_clusters):
    cluster_indices = [index for index, val in enumerate(color_values) if val == cluster]
    cluster_x = [x_int_values[i] for i in cluster_indices]
    cluster_y = [y_int_values[i] for i in cluster_indices]
    
    # Use 'color' parameter to set color for each cluster
    ax.scatter(cluster_x, cluster_y, label=f'Cluster {cluster:.0f}', color=color_list[i % len(color_list)])

# Replace integer ticks with string labels
plt.xticks(range(len(x_categories)), x_categories, rotation=45)
plt.yticks(range(len(y_categories)), y_categories)

plt.xlabel('Target Group')
plt.ylabel('Genre')

# Add the legend
plt.legend(title='Clusters', loc='upper left', bbox_to_anchor=(1, 1))

plt.grid(alpha=0.2)
plt.title("Netflix Content Clusters")
plt.tight_layout()

plt.show()
```
